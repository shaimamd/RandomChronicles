```python
import pandas as pd
from  matplotlib import pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("images_analyzed_productivity1.csv")
```

```python
# sizes=df["Productivity"].value_counts(sort=1)
# print(sizes)
```

Predicting the productivity based on independent variable, so its the time of the day here. Images analysed can be used as metric, but proiductivity gets to practice if the data is not metric

```python
###Dropping irrelevanmt columns, 80% its about organising the data!
df.drop(["Images_Analyzed"],axis=1, inplace=True)
##inplace updates the data frame
df.drop(["User"],axis=1, inplace=True)
```

```python
##Handle missing values, drop the rows or colums if one of the entries is blank
#df=df.dropna()
```

```python
##Non numeric data conversion to numeric
##Good=1, Bad=2
df.Productivity[df.Productivity =='Good']=1
df.Productivity[df.Productivity =='Bad']=2
```

```python
print(df.head())
```

```python
##Define dependent variables
Y=df["Productivity"].values
Y=Y.astype('int')
```

```python
##Define independent variables (Time, Coffee and age), just do this by dropping productivity
X=df.drop(labels=["Productivity"],axis=1)
```

```python
#Split the data into training and testing set, only way to test the accuracy of the algorithm
from sklearn.model_selection import train_test_split
##Pass the X and Y values to split, with 20% data is assigned to the test dataset, random_state=20, the randomness to be same when one runs again the same code
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2,random_state=20)
#print(X_train)
#print(X_test)
```

```python
##Random Forest is ensemble type of classifier, this also has regresser (RandomForestClassifier) which is a floating point number
##For regressor its a value, in this case its binary, it could be many but thisd case is classifier
###Use  RandomForestClassifier: for example No of images analysed or financial problems.
##Random forest also has randomness keep it same
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=10,random_state=30)
model.fit(X_train,Y_train)
```

```python
###Predict from the test dataset and compare from the original value
prediction_test=model.predict(X_test)
#print(prediction_test)
```

```python
from sklearn import metrics
print("Accuracy=",metrics.accuracy_score(Y_test,prediction_test))
```

```python
##Random Forest gives what parameters contributes the best, so one can extract feature importances
##Time, Coffee and Age are the factors,based on this which one is important?
#print(model.feature_importances_)
feature_list=list(X.columns)
feature_imp=pd.Series(model.feature_importances_,index=feature_list).sort_values(ascending=False)
print(feature_imp)
```
